```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from datetime import datetime, timedelta
# from pybit.unified_trading import HTTP
import warnings
from sklearn.ensemble import IsolationForest
```

```python
def load_train_test(ticker, train_size=1):
    '''
    Загрузка исторических данных с локального комп и разделение на трейн и тест
    '''
    df = pd.read_parquet(f"../hw2/data/{ticker}.parquet", engine='pyarrow')
    ind = int(len(df.index)*train_size)

    return df.iloc[:ind], df.iloc[ind:]
```

### 1. Поиск сдвигов и аномалий
Я не думаю что стоит удалять выбросы или сдвиги в истории цены актива, но этот пункт есть в задании, так что я взял самую простую библиотеку которая ищет сразу все виды аномалий, если находит то стави -1, удалять аномалии не буду, просто добавлю как признак

```python
df,_ = load_train_test('DOTUSDT')
```

```python
model = IsolationForest(contamination=0.1)
df['anomaly'] = model.fit_predict(df)
```

### 2. Фичиинжениринг 
Я нашёл на kagle уже готовый класс где рассчитываются все основные фичи для анализа поведения цены, и пременил их к своим данным

```python
from feature_engineering import Features
```

```python
feature_engineer = Features(df)
```

```python
# Apply all the feature engineering methods
asset_df = feature_engineer.price_based_features()
asset_df = feature_engineer.volume_based_features()
asset_df = feature_engineer.volatility_features(window=15)
asset_df = feature_engineer.momentum_features(window=15)
asset_df = feature_engineer.trend_features(window=14)
asset_df = feature_engineer.lag_rolling_features(lags=[1, 2, 3], windows=[5, 10, 20])
asset_df = feature_engineer.statistical_features(window=20)
asset_df = feature_engineer.time_based_features()
asset_df = feature_engineer.fill(True, True)  # ffill and bfill
```
